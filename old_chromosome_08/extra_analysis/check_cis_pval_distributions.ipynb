# Uncorrected

## 8p cis effects

```python
ttest_results_file = "8p_ciseffects_ttest_uncorrected.tsv"

ttest_results = pd.\
read_csv(ttest_results_file, sep="\t").\
rename(columns={"Name": "protein"}).\
set_index("protein")

pvals = ttest_results.\
loc[:, ttest_results.columns.str.endswith("pvalue")].\
stack().\
reset_index().\
rename(columns={0: "pvals"})

p = alt.Chart(pvals).mark_bar().encode(
    x=alt.X(
        "pvals",
        bin=alt.Bin(step=0.05)
    ),
    y=alt.Y(
        "count()"
    )
).properties(
    title="8p loss"
)
```

## 8q cis effects

```python
ttest_results_file = "8q_ciseffects_ttest_uncorrected.tsv"

ttest_results = pd.\
read_csv(ttest_results_file, sep="\t").\
rename(columns={"Name": "protein"}).\
set_index("protein")

pvals = ttest_results.\
loc[:, ttest_results.columns.str.endswith("pvalue")].\
stack().\
reset_index().\
rename(columns={0: "pvals"})

q = alt.Chart(pvals).mark_bar().encode(
    x=alt.X(
        "pvals",
        bin=alt.Bin(step=0.05)
    ),
    y=alt.Y(
        "count()"
    )
).properties(
    title="8q gain"
)
```

```python
alt.hconcat(p, q).properties(
    title="Cis effects uncorrected P value distributions"
).configure_title(
    anchor="middle"
)
```
