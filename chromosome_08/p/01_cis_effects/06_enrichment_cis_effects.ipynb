# Cis effects enrichment: Enrichr and g:Profiler

```python
import cnvutils
import os
```

```python
params = cnvutils.load_params(os.path.join("..", "data", "params.json"))
CHROMOSOME = params["CHROMOSOME"]
ARM = params["ARM"]
CIS_OR_TRANS = "cis"
```

## Enrichr

We will use Enrichr to look for pathways enriched for proteins that came up across multiple cancers in the cis effects analysis.

```python
cnvutils.run_enrichr(os.path.join("..", "data", f"common_summary_{CHROMOSOME}{ARM}_{CIS_OR_TRANS}.tsv"))
```

# g:Profiler

We will now use the [g:Profiler tool](https://biit.cs.ut.ee/gprofiler/gost) to look for pathways enriched for proteins that came up across multiple cancers in the cis effects analysis.

```python
cnvutils.run_gprofiler(os.path.join("..", "data", f"common_summary_{CHROMOSOME}{ARM}_{CIS_OR_TRANS}.tsv"))
```
