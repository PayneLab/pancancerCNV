```python
luad = cptac.Luad()
brca = cptac.Brca()
ovarian = cptac.Ovarian()
colon = cptac.Colon()
lscc = cptac.Lscc()
hnscc = cptac.Hnscc()
```

```python
luad_has_event = pd.read_csv("luad_has_event.csv")
brca_has_event = pd.read_csv("brca_has_event.csv")
ovarian_has_event = pd.read_csv("ovarian_has_event.csv")
colon_has_event = pd.read_csv("colon_has_event.csv")
lscc_has_event = pd.read_csv("lscc_has_event.csv")
hnscc_has_event = pd.read_csv("hnscc_has_event.csv")
```

```python
luad_has_event
```

```python
table = pd.crosstab(hnscc_has_event['gain_event'], hnscc_has_event['loss_event'])
oddsratio, pvalue = stats.fisher_exact(table)

print(oddsratio)
print(pvalue)
```

```python
table
```

```python
labelizer = lambda k: {('0','0'): f"No Event \nn={table.loc[int(k[0]), int(k[1])]}", ('0','1'): f"Loss Event \nn={table.loc[int(k[0]), int(k[1])]}", ('1','1'): f"Both Events\nn={table.loc[int(k[0]), int(k[1])]}", ('1','0'): f"Gain Event \nn={table.loc[int(k[0]), int(k[1])]}"}[k]
mosaic(hnscc_has_event, ['gain_event', 'loss_event'], labelizer=labelizer, axes_label=False, title="HNSCC")
plt.text(0.2, -0.1, f"Odds Ratios: {oddsratio}")
plt.text(0.2, -0.2, f"P Value: {pvalue}")
```
