# Make chromosome plots

Plot the net amount and type of CNV event at each gene in the chromosome. By net amount we mean the number of patients with a gain at a gene, minus the number of patients with a loss at that same gene. So if 15 patients have gained a gene and 75 patients have lost it, it would be counted as an overall lost gene, and displayed with an intensity of -60.

```python
import cnvutils
from IPython.display import Image

CHROMOSOME = 8
```

## Gradient plots

In these plots, all counts are divided by the total number of patients in the cohort in order to yield proportions.

```python
Image(cnvutils.make_chr_gradient_plot(
    source="cptac",
    chromosome=CHROMOSOME,
))
```

```python
Image(cnvutils.make_chr_gradient_plot(
    source="gistic",
    chromosome=CHROMOSOME,
    level="gene",
))
```

## Line plots

These plots currently use absolute counts, not proportions.

```python
Image(cnvutils.make_chr_line_plot(
    source="cptac",
    chromosome=CHROMOSOME,
))
```

```python
Image(cnvutils.make_chr_line_plot(
    source="gistic",
    level="gene",
    chromosome=CHROMOSOME,
))
```
