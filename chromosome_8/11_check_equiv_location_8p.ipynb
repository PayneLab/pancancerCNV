# Check the location of proteins that pass as equivalently expressed between samples with and without the event

If they're close to each other, that may indicate regulation of transcription to ameliorate the effects of CNVs.

```python
import pandas as pd
import numpy as np
import altair as alt
import cnvutils as ut
```

```python
CHROMOSOME = "8"
ARM = "p"
CIS_OR_TRANS = "cis"
```

```python
equiv = pd.read_csv(f"pancancer_summary_equiv_{CHROMOSOME}{ARM}_{CIS_OR_TRANS}.tsv", sep="\t")
```

```python
ut.make_chromosome_plot(chromo=CHROMOSOME, arm=ARM, genes=equiv["protein"])
```

```python
top_equiv = equiv[equiv["num_cancers"] > 3]
ut.make_chromosome_plot(chromo=CHROMOSOME, arm=ARM, genes=top_equiv["protein"])
```
