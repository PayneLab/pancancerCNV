# Preparing 8p cis effects data for rSEA

## Setup

```python
import pandas as pd
import numpy as np
import os
```

```python
ttest_results_file = "8p_ciseffects_ttest.tsv"

ttest_results = pd.\
read_csv(ttest_results_file, sep="\t").\
rename(columns={"Name": "protein"})
```

```python
ttest_results
```

## Find how many cancers each protein was different in

```python
# We want to get our table to have these columns:
# cancer_type, protein, change, p_value

# Slice out the info for each cancer type and reshape individually

ttest_results = ttest_results.melt(
    id_vars="protein",
    var_name="variable",
    value_name="value"
)
```

```python
split_vars = ttest_results["variable"].str.split("_", n=1, expand=True)

ttest_results = ttest_results.assign(
    cancer_type=split_vars[0],
    variable=split_vars[1]
)
```

```python
ttest_results
```

```python
ttest_results.\
drop_duplicates(keep="first").\
set_index(["cancer_type", "protein", "variable"])
```
