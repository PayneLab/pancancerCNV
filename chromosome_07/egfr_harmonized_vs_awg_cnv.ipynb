```python
import altair as alt
import os
import pandas as pd
```

```python
awg = pd.read_csv(os.path.join("data", "cnv_counts.tsv"), sep='\t')
```

```python
awg
```

```python
awg_e = awg.drop(columns=["Database_ID"])
df = awg_e[awg_e["Name"] == "EGFR"].drop_duplicates(keep="first")
df = df.assign(value_prop=df["value"] / df["cancer_type_total_patients"])
df
```

```python
def make_plot(df, event_type):

    df = df[df["variable"] == event_type]

    return alt.Chart(df).mark_bar().encode(
        x=alt.X(
            "cancer:N",
            axis=alt.Axis(
                title=None,
            ),
        ),
        y=alt.Y(
            "value_prop",
            axis=alt.Axis(
                title=event_type
            ),
            scale=alt.Scale(
                domain=(0, 1)
            )
        ),
    )

alt.vconcat(
    make_plot(df, "gain"),
    make_plot(df, "loss"),
)
```
