# Check the location of cis proteins that pass as equivalently expressed between samples with and without the event

If they're close to each other, that may indicate regulation of transcription to ameliorate the effects of CNVs.

```python
import os
import pandas as pd
import numpy as np
import altair as alt
import cnvutils as ut
```

```python
# These variables specify which chromosome and arm we're working on, and whether to do cis or trans effects
params = ut.load_params(os.path.join("..", "data", "params.json"))
CHROMOSOME = params["CHROMOSOME"]
ARM = params["ARM"]
CIS_OR_TRANS = "cis"
```

```python
equiv = pd.read_csv(os.path.join("..", "data", f"{CHROMOSOME}{ARM}_{CIS_OR_TRANS}_common_equiv.tsv"), sep="\t")
```

```python
ut.make_chromosome_plot(chromo=CHROMOSOME, arm=ARM, genes=equiv["protein"].tolist())
```

```python
top_equiv = equiv[equiv["num_cancers"] > 3]
ut.make_chromosome_plot(chromo=CHROMOSOME, arm=ARM, genes=top_equiv["protein"].tolist())
```

## Compare to the locations of the significantly different proteins

```python
diff = pd.read_csv(os.path.join("..", "data", f"{CHROMOSOME}{ARM}_{CIS_OR_TRANS}_common_diff.tsv"), sep="\t")
```

```python
ut.make_chromosome_plot(chromo=CHROMOSOME, arm=ARM, genes=diff["protein"].tolist())
```

```python
top_diff = diff[diff["num_cancers"] > 3]
ut.make_chromosome_plot(chromo=CHROMOSOME, arm=ARM, genes=top_diff["protein"].tolist())
```
