```python
import cptac
import numpy as np
import pandas as pd
import cnvutils
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
luad = cptac.Luad()
luad_cnv = luad.get_CNV()
```

```python
locations = cnvutils.get_gene_locations()
```

```python
luad_cnv = luad_cnv.transpose()
patients = luad_cnv.columns
luad_cnv_loc = luad_cnv.join(locations)
```

```python
luad_cnv_8 = luad_cnv_loc[luad_cnv_loc.chromosome == '8']
```

```python
patients_of_interest = ['C3L-00080','C3N-02155', 'C3L-01632', 'C3L-02345', 'C3N-00169', 'C3N-00704', 'C3N-01021', 'C3N-01030', 'C3N-01415', 'C3N-02000', 'C3N-02002', 'C3N-02067', 'C3N-02145', 'C3N-02149', 'C3N-02433']
```

```python
for patient in list(patients):
    plt.scatter(x='start_bp', y=patient, data=luad_cnv_8)
    plt.ylim(-.5, 3)
    plt.title(patient)
    plt.show()
```

```python
for patient in list(patients_of_interest):
    plt.scatter(x='start_bp', y=patient, data=luad_cnv_8)
    low = luad_cnv_8[luad_cnv_8[patient] < 0]
    high = luad_cnv_8[luad_cnv_8[patient] > 0]
    print(high[['start_bp']].sort_values('start_bp'))
    plt.ylim(-.5, .5)
    plt.title(patient)
    plt.show()
```
