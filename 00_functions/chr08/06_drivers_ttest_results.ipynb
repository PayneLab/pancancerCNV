```python
import pandas as pd

df = pd.DataFrame({"a": [1, 1, 1], "b": [2, 3, 4]})

def helper(x, i):
    x["plus"] = x["a"] + i
    x["mult"] = x["a"] * i
    x["exp"] = x["a"] ** i
    return x

df.apply(lambda x: helper(x, 3), axis=1)
```

# Driver gene plots

```python
import cnvutils
import IPython.display
```

```python
paths = cnvutils.multi_runner(
    func=cnvutils.make_drivers_manhattan_plot,
    sources=["cptac", "gistic"],
    levels=["gene"],
    chromosomes_events={
        8: {
            "p": ["loss"],
            "q": ["gain"],
        },
    },
    more_dicts=[
        {
            "name": "cis_or_trans",
            "vals": ["cis", "trans"],
        },
        {
            "name": "proteomics_or_transcriptomics",
            "vals": ["transcriptomics", "proteomics"],
        },
    ]
)
```

```python
cis_paths = [path for path in paths if "cis" in path]
```

```python
trans_paths = [path for path in paths if "cis" in path]
```

```python
IPython.display.display(*[IPython.display.Image(path) for path in cis_paths])
```

```python
IPython.display.display(*[IPython.display.Image(path) for path in trans_paths])
```
