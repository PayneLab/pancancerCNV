# Make chromosome plots

```python
import cnvutils
from IPython.display import Image

CHROMOSOME = 8
```

## Gradient plots

```python
Image(cnvutils.make_chr_gradient_plot(
    source="cptac",
    chromosome=CHROMOSOME,
))
```

```python
Image(cnvutils.make_chr_gradient_plot(
    source="gistic",
    chromosome=CHROMOSOME,
    level="gene",
))
```

## Line plots

```python
Image(cnvutils.make_chr_line_plot(
    source="cptac",
    chromosome=CHROMOSOME,
))
```

```python
Image(cnvutils.make_chr_line_plot(
    source="gistic",
    level="gene",
    chromosome=CHROMOSOME,
))
```
