## 8p and 8q event relation

Here we run a series of fisher tests to determine if the two events (loss of 8p and gain of 8q) are coorelated. 

```python
import cptac
import pandas as pd
import scipy.stats as stats
from statsmodels.graphics.mosaicplot import mosaic
import matplotlib.pyplot as plt
```

## Import Data

```python
luad_has_event = pd.read_csv("../luad_has_event.csv")
brca_has_event = pd.read_csv("../brca_has_event.csv")
ovarian_has_event = pd.read_csv("../ovarian_has_event.csv")
colon_has_event = pd.read_csv("../colon_has_event.csv")
lscc_has_event = pd.read_csv("../lscc_has_event.csv")
hnscc_has_event = pd.read_csv("../hnscc_has_event.csv")
```

```python
luad_table = pd.crosstab(luad_has_event['gain_event'], luad_has_event['loss_event'])
brca_table = pd.crosstab(brca_has_event['gain_event'], brca_has_event['loss_event'])
ovarian_table = pd.crosstab(ovarian_has_event['gain_event'], ovarian_has_event['loss_event'])
colon_table = pd.crosstab(colon_has_event['gain_event'], colon_has_event['loss_event'])
lscc_table = pd.crosstab(lscc_has_event['gain_event'], lscc_has_event['loss_event'])
hnscc_table = pd.crosstab(hnscc_has_event['gain_event'], hnscc_has_event['loss_event'])
```

```python
luad_odds, luad_pvalue = stats.fisher_exact(luad_table)
brca_odds, brca_pvalue = stats.fisher_exact(brca_table)
ovarian_odds, ovarian_pvalue = stats.fisher_exact(ovarian_table)
colon_odds, colon_pvalue = stats.fisher_exact(colon_table)
lscc_odds, lscc_pvalue = stats.fisher_exact(lscc_table)
hnscc_odds, hnscc_pvalue = stats.fisher_exact(hnscc_table)
```

```python
luad_labelizer = lambda k: {('0','0'): f"No Event \nn={luad_table.loc[int(k[0]), int(k[1])]}", ('0','1'): f"Loss Event \nn={luad_table.loc[int(k[0]), int(k[1])]}", ('1','1'): f"Both Events\nn={luad_table.loc[int(k[0]), int(k[1])]}", ('1','0'): f"Gain Event \nn={luad_table.loc[int(k[0]), int(k[1])]}"}[k]
brca_labelizer = lambda k: {('0','0'): f"No Event \nn={brca_table.loc[int(k[0]), int(k[1])]}", ('0','1'): f"Loss Event \nn={brca_table.loc[int(k[0]), int(k[1])]}", ('1','1'): f"Both Events\nn={brca_table.loc[int(k[0]), int(k[1])]}", ('1','0'): f"Gain Event \nn={brca_table.loc[int(k[0]), int(k[1])]}"}[k]
ovarian_labelizer = lambda k: {('0','0'): f"No Event \nn={ovarian_table.loc[int(k[0]), int(k[1])]}", ('0','1'): f"Loss Event \nn={ovarian_table.loc[int(k[0]), int(k[1])]}", ('1','1'): f"Both Events\nn={ovarian_table.loc[int(k[0]), int(k[1])]}", ('1','0'): f"Gain Event \nn={ovarian_table.loc[int(k[0]), int(k[1])]}"}[k]
colon_labelizer = lambda k: {('0','0'): f"No Event \nn={colon_table.loc[int(k[0]), int(k[1])]}", ('0','1'): f"Loss Event \nn={colon_table.loc[int(k[0]), int(k[1])]}", ('1','1'): f"Both Events\nn={colon_table.loc[int(k[0]), int(k[1])]}", ('1','0'): f"Gain Event \nn={colon_table.loc[int(k[0]), int(k[1])]}"}[k]
lscc_labelizer = lambda k: {('0','0'): f"No Event \nn={lscc_table.loc[int(k[0]), int(k[1])]}", ('0','1'): f"Loss Event \nn={lscc_table.loc[int(k[0]), int(k[1])]}", ('1','1'): f"Both Events\nn={lscc_table.loc[int(k[0]), int(k[1])]}", ('1','0'): f"Gain Event \nn={lscc_table.loc[int(k[0]), int(k[1])]}"}[k]
hnscc_labelizer = lambda k: {('0','0'): f"No Event \nn={hnscc_table.loc[int(k[0]), int(k[1])]}", ('0','1'): f"Loss Event \nn={hnscc_table.loc[int(k[0]), int(k[1])]}", ('1','1'): f"Both Events\nn={hnscc_table.loc[int(k[0]), int(k[1])]}", ('1','0'): f"Gain Event \nn={hnscc_table.loc[int(k[0]), int(k[1])]}"}[k]
```

```python
mosaic(brca_has_event, ['gain_event', 'loss_event'], labelizer=brca_labelizer, axes_label=False, title="BRCA")
plt.text(0.2, -0.1, f"Odds Ratios: {brca_odds}")
plt.text(0.2, -0.2, f"P Value: {brca_pvalue}")
```
